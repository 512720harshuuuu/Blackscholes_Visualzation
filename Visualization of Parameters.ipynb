# Find Implied Volatility 

```python
def CallBS(T,S_0,K,r,sigma):
    a = (np.log(S_0/K))
    b = (r+np.square(sigma)/2)*T
    c = sigma*np.sqrt(T)
    dplus =  (a + b)/c
    dminus = dplus - sigma*(np.sqrt(T))
    return (S_0*norm.cdf(dplus, loc=0, scale=1)- K*math.exp(-r*T)*norm.cdf(dminus, loc=0, scale=1))
def dplus(T,S_0,K,r,sigma):
    a = (np.log(S_0/K))
    b = (r+np.square(sigma)/2)*T
    c = sigma*np.sqrt(T)
    dplus =  (a + b)/c
    dminus = dplus - sigma*(np.sqrt(T))
    return dplus
```

```python
import math

from scipy.stats import norm
def findsigma(T,S_0,K,r,Ck):
    sigma = [Ck/(0.4*S_0*np.sqrt(T))]
    for i in range(1000):
        gdash = S_0*(1/np.sqrt(2*math.pi))*math.exp(-np.square(dplus(T,S_0,K,r,sigma[i]))/2)*np.sqrt(T)
        sigma.append(sigma[i] - ((CallBS(T,S_0,K,r,sigma[i]) - Ck)/(gdash)))
    return sigma
```
